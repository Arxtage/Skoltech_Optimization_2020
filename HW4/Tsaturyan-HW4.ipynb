```python
import matplotlib.pyplot as plt
import numpy as np

def obective_function(x):
    return(x**2+1)

def lagrangian(x,lda):
    return((1+lda)*x**2-6*lda*x+(1+8*lda))

lambdas=[-2,-1,-0.5,1,2,3,4]
x = np.linspace(-1, 6, 100)

plt.figure(figsize=(12,10))
for l in lambdas:
    plt.plot(x, lagrangian(x, l), label ="lambda=" +str(l))
    plt.legend()
plt.plot(x, obective_function(x), label ='Objective Function', linestyle='dashed')

plt.plot(2,5,'ro')
plt.legend()
plt.title('Curves for different lambdas',size=20)
plt.show()
```

To use the theorem [source: https://ljk.imag.fr/membres/Anatoli.Iouditski/cours/convex/chapitre_4.pdf] let us introduce new veariable:
$$t=-P^{-1}(A^T\lambda-q)$$

Then the dual problem is rewrited as:
$$-\frac{1}{2}t^TPt+b^T\lambda \to \max | A^T\lambda+Pt=q, \lambda \geqslant 0$$

**In our case:**
All the requirements of theorem are fulfiled: $P$ is symmetric positive definite, primal objective at x is equal to the dual objective at $(\lambda,t)$ 
So, both primal and dual problems are solvable and the pair $(x;(\lambda,t))$ of feasible solutions to the problem is comprised of optimal solutions to them.
